import requests
raw_data = requests.get('https://github.com/topics/javascript').text
raw_data

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(raw_data,'lxml')
soup
```

```python
first_div = soup.find('article',class_='border rounded color-shadow-small color-bg-subtle my-4')
first_div
```

```python
repo_name = first_div.find('a',class_='text-bold wb-break-word').text.strip()
repo_name
```

```python
username_str=''
username=first_div.find('h3',class_='f3 color-fg-muted text-normal lh-condensed').text.strip()
username_str+= username
username_str_split=username_str.split()
user_name=username_str_split[0]
user_name=user_name.strip()
user_name
```

```python
stars = first_div.find('span',class_='Counter js-social-count').text
stars
```

```python
base_url = 'https://github.com/'
url = first_div.find('a', class_ = 'text-bold wb-break-word')['href']
repos_url = base_url + url
repos_url
```

```python
all_divs = soup.find_all('article',class_='border rounded color-shadow-small color-bg-subtle my-4')
all_divs
```

```python
common_load_url = 'https://github.com/topics/javascript?page='
load_lst = []
for num in range(1,8):
    load_lst.append(common_load_url+str(num))
load_lst
```

```python
base_url = 'https://github.com'
all_data_lst = []
repoNo=1
for item in load_lst:
    all_raw_data = requests.get(item).text
    all_soup = BeautifulSoup(all_raw_data,'lxml')
    all_divs = all_soup.find_all('article' ,class_='border rounded color-shadow-small color-bg-subtle my-4')
    for data in all_divs:
        if(repoNo>200):
          continue
        else:
          repository_name = data.find('a',class_='text-bold wb-break-word').text.strip()
          username_str=''
          username=data.find('h3',class_='f3 color-fg-muted text-normal lh-condensed').text.strip()
          username_str+= username
          username_str_split=username_str.split()
          user_name=username_str_split[0]
          user_name=user_name.strip()
          stars = data.find('span',class_='Counter js-social-count').text
          url = data.find('a', class_ = 'text-bold wb-break-word')['href']
          repos_url = base_url + url  
        
       
        
          all_data_dict = {
            'Sr No':repoNo,
            'Repository Name':repository_name,
            'User Name ':user_name,
            'Stars ':stars,
            'URL':repos_url
          }
        
          all_data_lst.append(all_data_dict) 
          repoNo+=1
    
print(all_data_lst)  
```

```python
import pandas as pd
all_data = pd.DataFrame(all_data_lst)
```
